```python
from  matplotlib import pyplot as plt
%matplotlib inline
```

## bitwise 操作

```python
import numpy as np 
import cv2

rectangle = np.zeros((300,300),dtype="uint8")
cv2.rectangle(rectangle,(25,25),(275,275),255,-1)
plt.imshow(rectangle)

circle = np.zeros((300,300),dtype="uint8")
cv2.circle(circle,(150,150),150,255,-1)
plt.imshow(circle)

bitwiseAnd = cv2.bitwise_and(rectangle,circle)
plt.imshow(bitwiseAnd)


bitwiseOr = cv2.bitwise_or(rectangle,circle)
plt.imshow(bitwiseOr)


bitwiseXor = cv2.bitwise_xor(rectangle,circle)
plt.imshow(bitwiseXor)

bitwiseNot = cv2.bitwise_not(rectangle)
plt.imshow(bitwiseNot)
```

## 图像融合

```python
# 底板图案
bottom_pic = '../images/dog.jpg'
# 上层图案
top_pic = '../images/cat.jpg'

import cv2
bottom = cv2.imread(bottom_pic)
top = cv2.imread(top_pic)
bottom_size = bottom.shape
top_size = top.shape
common_size = (max(bottom_size[0], top_size[0]), max(bottom_size[1], top_size[1]))
bottom = cv2.resize(bottom, common_size)
top = cv2.resize(top, common_size)
# 权重越大，透明度越低
overlapping = cv2.addWeighted(bottom, 0.5, top, 0.5, 0)
# 保存叠加后的图片
plt.imshow(overlapping)
```
